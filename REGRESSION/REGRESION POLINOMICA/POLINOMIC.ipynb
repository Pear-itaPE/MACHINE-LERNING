```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd 
from sklearn.linear_model import LinearRegression
```

```python
#Cargamos el dataset
dp= pd.read_csv("Position_Salaries.csv")
```

```python
x= dp.iloc[:, 1:2].values #Vector a matriz
y= dp.iloc[:, 2].values
```

```python
#Dividor el dataset en conjunto de entrenamiento y conjunto de testing
"""
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split (x,y, test_size=0.2, random_state=0
"""
```

```python
#Ajustar egresion lineal
lin_reg= LinearRegression()
lin_reg.fit(x,y)
```

```python
#Ajustar regresion polinomica
from sklearn.preprocessing import PolynomialFeatures
poly_reg= PolynomialFeatures(degree=4)
X_poly = poly_reg.fit_transform(x)
lin_reg2 = LinearRegression()
lin_reg2.fit(X_poly,y)
```

```python
#Visualziacion de los resultados del modelo regresion lineal
plt.scatter(x, y, color="red")
plt.plot(x, lin_reg.predict(x), color="blue") 
plt.title("Modelo de Regresión Lineal")
plt.xlabel("Posición del empleado")
plt.ylabel("Sueldo (en millones de $)")
plt.ticklabel_format(style="plain", axis="y")
plt.show()
```

```python
#Visualziacion de los resultados del modelo polinomico
plt.scatter(x, y, color="red")
plt.plot(x, lin_reg2.predict(X_poly), color="blue") 
plt.title("Modelo de Regresión Lineal")
plt.xlabel("Posición del empleado")
plt.ylabel("Sueldo (en millones de $)")
plt.ticklabel_format(style='plain', axis='y') 
plt.show()
```

```python
#Visualziacion de los resultados del modelo polinomico
#Ajustando linea
X_grid = np.arange(x.min(), x.max(), 0.1)
X_grid = X_grid.reshape(len(X_grid), 1)

plt.scatter(x, y, color="red")
plt.plot(X_grid, lin_reg2.predict(poly_reg.fit_transform(X_grid)), color="blue") 

plt.title("Modelo de Regresión Polinómica")
plt.xlabel("Posición del empleado")
plt.ylabel("Sueldo (en millones de $)")
plt.ticklabel_format(style='plain', axis='y')
plt.show()
```

```python
#Prediccion de nuestros modelos (Regresion Lineal)
lin_reg.predict([[6.5]])
```

```python
#Prediccion de nuestros modelos (Polinomico)
lin_reg2.predict(poly_reg.fit_transform([[6.5]]))
```
