```python
#REGRESION LINEAL MULTIPLE
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression  
from sklearn.metrics import classification_report, accuracy_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn import preprocessing
```

```python
#Cargar el Dataset
lm= pd.read_csv("50_Startups.csv")
```

```python
#Definir x/y
x= lm.iloc[:, :-1].values
y= lm.iloc[:, 4].values
```

```python
# CODIFICAR DATOS CATEGORICOS
le_x = preprocessing.LabelEncoder()
x[:, 3] = le_x.fit_transform(x[:, 3])
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [3])], remainder='passthrough')
X = np.array(ct.fit_transform(x))

# Evitar la trampa de las variable ficticias
X = X[:, 1:]
```

```python
#Dividir el data set en conjunto de entrenamiento y conjunto de testing
X_train, X_test, y_train, y_test = train_test_split( X,y, test_size=0.2, random_state=0)
```

```python
#Ajustar el modelo de regresion multiple con el conjunto de entrenamiento
regression= LinearRegression()
regression.fit(X_train, y_train) 
```

```python
#Prediccion de los resultados con los conjuntos de testing
y_pred = regression.predict(X_test)
```
