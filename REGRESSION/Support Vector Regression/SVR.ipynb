```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd 
from sklearn.linear_model import LinearRegression
```

```python
#Cargamos el dataset
dp= pd.read_csv("Position_Salaries.csv")
```

```python
X= dp.iloc[:, 1:2].values 
y= dp.iloc[:, 2].values
```

```python
"""#Dividor el dataset en conjunto de entrenamiento y conjunto de testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split (x,y, test_size=0.2, random_state=0)"""
```

```python
#Escalado de variables
from sklearn.preprocessing import StandardScaler
sc_X=StandardScaler()
sc_y=StandardScaler()
X=sc_X.fit_transform(X)
y=sc_y.fit_transform(y.reshape(-1,1))
```

```python
#Ajustar regresion con el dataset
from sklearn.svm import SVR 
regression = SVR (kernel= "rbf")
regression.fit(X,y)
```

```python
# Realizar la predicción
y_pred = regression.predict(sc_X.transform([[6.5]]))

# Reformar y_pred a un array bidimensional
y_pred_reshaped = y_pred.reshape(-1, 1)

# Aplicar la transformación inversa
y_pred_inverse = sc_y.inverse_transform(y_pred_reshaped)
```

```python
#Visualizacion de los resultados VSR

X_grid = np.arange(X.min(), X.max(), 0.1)
X_grid = X_grid.reshape(len(X_grid), 1)

plt.scatter(X, y, color="red")
plt.plot(X, regression.predict(X), color="blue") 

plt.title(" MODELO DE REGRESION SVR")
plt.xlabel("POSICION DEL EMPLEADO ")
plt.ylabel("SUELDO EN $ ")
plt.ticklabel_format(style='plain', axis='y')
plt.show()
```
